# Question 1

# Write a Python program that takes the age of a person as input and classifies them into one of the following age groups:

# "Child" if the age is between 0 and 12 (inclusive).
# "Teenager" if the age is between 13 and 19 (inclusive).
# "Adult" if the age is between 20 and 64 (inclusive).
# "Senior" if the age is 65 or older.

```python

age = int(input("Enter your age: "))

if age < 0:
    print("Invalid age entered.")
elif age <= 12:
    print("Child")
elif age <= 19:
    print("Teenager")
elif age <= 64:
    print("Adult")
else:
    print("Senior")
```

# Question 2

# Write a Python program that takes three numbers as input and determines the largest among them. Also, check if any or all of the numbers are equal.

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number: "))

# Find the largest number
if num1 >= num2 and num1 >= num3:
    largest = num1
elif num2 >= num1 and num2 >= num3:
    largest = num2
else:
    largest = num3

print("The largest number is:", largest)

# Check equality
if num1 == num2 == num3:
    print("All three numbers are equal.")
elif num1 == num2 or num1 == num3 or num2 == num3:
    print("Two numbers are equal.")
else:
    print("All numbers are different.")
```

# Question 4

# Write a Python program to store and display the details of a book.

# The program should ask the user to input:

# Title of the book
# Author's name
# Year of publication
# Price of the book

# Store the entered details in a tuple and display them.

```python
title = input("Enter book title: ")
author = input("Enter author's name: ")
year = int(input("Enter year of publication: "))
price = float(input("Enter book price: "))

book = (title, author, year, price)

print("\nBook Details")
print("Title:", book[0])
print("Author:", book[1])
print("Year of Publication:", book[2])
print("Price: $", book[3])
```

# Question 5

# Write a program to check if a user's chosen subject is available in the list of offered subjects.

```python
subjects = ["Math", "Science", "English", "Computer", "Art", "Physics"]

subject = input("Enter the subject you are interested in: ")

if subject.lower() in [s.lower() for s in subjects]:
    print(subject, "is available.")
else:
    print(subject, "is not available.")
```

# Question 7

# Write a Python program that simulates a simple login system.

# The correct username is 'admin' and the correct password is 'admin@123'.

# The program should take username and password as input from the user.

# It should check the input and display messages according to the following conditions:

# If both username and password are correct → print "Login Successful".
# If username is wrong but password is correct → print "Invalid Username".
# If username is correct but password is wrong → print "Invalid Password".
# If both username and password are wrong → print "Invalid Username and Password".

```python
username = input("Enter username: ")
password = input("Enter password: ")

if username == "admin" and password == "admin@123":
    print("Login Successful")
elif username != "admin" and password == "admin@123":
    print("Invalid Username")
elif username == "admin" and password != "admin@123":
    print("Invalid Password")
else:
    print("Invalid Username and Password")
```

# Question 8

# Write a Python program to check a person's eligibility for a loan based on age, credit score, and annual income.

```python
age = int(input("Enter your age: "))
credit_score = int(input("Enter your credit score: "))
income = float(input("Enter your annual income ($): "))

if age < 18:
    print("You are not eligible for the loan due to age.")
elif credit_score < 650:
    print("You are not eligible for the loan due to low credit score.")
elif credit_score >= 700 and income >= 25000:
    print("You are eligible for the loan!")
elif income >= 30000:
    print("You are eligible for the loan!")
else:
    print("You are not eligible for the loan due to insufficient income.")
```

# Question 9

# Write a Python program to check whether the entered letter is a vowel, a consonant, or invalid input.

# The program should take a single character as input.

# If the character is a vowel (a, e, i, o, u), print "Vowel".
# If it is an alphabet but not a vowel, print "Consonant".
# Otherwise, print "Please enter valid alphabet".

```python
letter = input("Enter a single letter: ")

if len(letter) == 1 and letter.isalpha():
    if letter.lower() in "aeiou":
        print(letter, "is a Vowel")
    else:
        print(letter, "is a Consonant")
else:
    print("Please enter valid alphabet")
```

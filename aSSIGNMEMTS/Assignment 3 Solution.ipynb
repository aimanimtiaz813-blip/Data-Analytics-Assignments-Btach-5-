# Question 2:

A customer orders several food items. Add the items to a list and then:
order = ["burger", "fries", "coke"]

Add a new item "ice cream" using append()
Remove an item "biryani" using remove()
Show the final order

```python
order = ["burger", "fries", "coke"]

order.append("ice cream")

if "biryani" in order:
    order.remove("biryani")

print(order)
```

## Question 4:

Track passengers on a bus:
passengers = ["Ali", "Zara", "Ahmed"]

Insert a new passenger "Sara" at position 2
Remove passenger who got off "Zara"
Reverse the order to show who entered first

```python
passengers = ["Ali", "Zara", "Ahmed"]

passengers.insert(2, "Sara")
passengers.remove("Zara")
passengers.reverse()

print(passengers)
```

### Question 5:

text = 'hello python world'

Replace spaces with underscores in the string.
Split the modified string.
Join the list elements with a hyphen (-) between them.
Print all three results.

```python
text = "hello python world"

modified = text.replace(" ", "_")
split_text = modified.split("_")
joined_text = "-".join(split_text)

print(modified)
print(split_text)
print(joined_text)
```

### Question 7:

You have a product code that follows the format "SKU-12345-XYZ".

Find the first position of the dash (-).
Count the number of dashes (-) in the product code.
Remove the "SKU-" prefix and "-XYZ" suffix from the product code, Extract the numeric part 12345.

```python
code = "SKU-12345-XYZ"

print(code.find("-"))
print(code.count("-"))

numeric_part = code.replace("SKU-", "").replace("-XYZ", "")
print(numeric_part)
```

## Question 9:

text = " Data Analysis with Python is very powerful "

Perform the following tasks:

Remove extra spaces using strip().
Convert the string into title case using title().
Replace "Powerful" with "Important".
Find the position of the word "Python".

```python
text = "    Data Analysis with Python is very powerful    "

text = text.strip()
text = text.title()

print(text.replace("Powerful", "Important"))
print(text.find("Python"))
```

## Question 10:

Given the list:
data = [12, 45, 2, 67, 34, 89, 23, 90]

Perform the following tasks:

Sort the list in ascending order.
Remove the highest value from the list.
Insert the value 50 at index 3.
Finally, reverse the updated list and display the result.

```python
data = [12, 45, 2, 67, 34, 89, 23, 90]

data.sort()
data.remove(max(data))
data.insert(3, 50)
data.reverse()

print(data)
```

# Write a Python program that asks the user to guess a secret word.

# Keep asking until the user enters the correct word ("python").
# When guessed correctly, print "🎉 You guessed it!"

```python
secret_word = "python"

while True:
    guess = input("Guess the word: ")

    if guess == secret_word:
        print("🎉 You guessed it!")
        break
```

# Question 2: Stop Sending Emails After 3 Users (Using for Loop and break)

# Write a Python program that sends emails to users.

# Stop sending emails after the first 3 users using the break statement.

```python

users = ["Ali", "Sara", "John", "Zara", "Mike"]

count = 0

for user in users:
    print("Email sent to", user)
    count += 1

    if count == 3:
        print("Limit reached. Stop sending emails.")
        break
```

# Question 3: Stop Monitoring Temperature if Critical Temperature is Reached (Using for Loop and break)

# Write a Python program that monitors temperatures.

# If the temperature reaches the critical value (40 or above), stop monitoring using the break statement.
# temps = [22, 25, 30, 45, 28]
# critical = 40

```python

temps = [22, 25, 30, 45, 28]
critical = 40

for temp in temps:
    if temp >= critical:
        print("Critical temperature reached! Stop monitoring.")
        break

    print("Temperature:", temp)
```

# Question 4: Calculate Total Price of Items in a Shopping Cart (Using for Loop)

# Write a Python program to calculate the total price of all items in the shopping cart.

# cart = [100, 250, 50, 300]

```python
cart = [100, 250, 50, 300]

total = 0

for price in cart:
    total += price

print("Total Price:", total)
```

# Question 5: Skip Customers with Unpaid Bills (Using for Loop and continue)

# Write a Python program that grants access only to customers who have paid their bills.

# Use the continue statement to skip customers with unpaid bills.
# customers = [
   #  {"name": "Ali", "paid": True},
   #  {"name": "Sara", "paid": False},
   #  {"name": "John", "paid": True}

```python
customers = [
    {"name": "Ali", "paid": True},
    {"name": "Sara", "paid": False},
    {"name": "John", "paid": True}
]

for customer in customers:
    if customer["paid"] == False:
        continue

    print("Access granted to:", customer["name"])
```

# Question 7: Extract Domain Names from Email List (Using for Loop)

# Write a Python program that extracts the domain names from the given email addresses using a for loop.

# emails = ["ali@gmail.com", "sara@yahoo.com", "john@hotmail.com"]

```python

emails = ["ali@gmail.com", "sara@yahoo.com", "john@hotmail.com"]

domains = []

for email in emails:
    domain = email.split("@")[1]
    domains.append(domain)

print("Domains:", domains)
```

# Question 8: Print Only Overdue Tasks (Using for Loop)

# Write a Python program that prints only the tasks whose status is "pending" using a for loop.

# tasks = [
   #  {"task": "Pay bills", "status": "done"},
   #  {"task": "Submit report", "status": "pending"},
   #  {"task": "Clean house", "status": "pending"}

```python

tasks = [
    {"task": "Pay bills", "status": "done"},
    {"task": "Submit report", "status": "pending"},
    {"task": "Clean house", "status": "pending"}
]

for task in tasks:
    if task["status"] == "pending":
        print("Overdue:", task["task"])
```

# Question 9: Stop the Loop When a Negative Number is Entered (Using while True and break)

# Write a Python program that keeps asking the user to enter numbers and prints them.

# The program should stop if the user enters a negative number.
# Use a while True loop with a break statement.

```python

while True:
    number = int(input("Enter a number (negative to stop): "))

    if number < 0:
        print("Negative number entered. Loop stopped!")
        break

    print("You entered:", number)
```

# Question 10: Skip Weekends While Printing Workdays (Using for Loop and continue)

# Write a Python program that prints only the workdays from the given list.

# Skip Saturday and Sunday using the continue statement.

```python

days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

for day in days:
    if day == "Saturday" or day == "Sunday":
        continue

    print("Workday:", day)
```
